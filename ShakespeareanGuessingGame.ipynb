```python
import random

# Generate a number between 0 and 50
actual_number = random.randint(0, 50)

# Define a list of Shakespearan quotes to guide the user
quotes = {
    "up":
     ["\nThe fault, dear Brutus, is not in our stars, But in ourselves, that we are underlings.", 
      "\nHe that climbs the tall tree has won right to the fruit.", 
      "\nAnd in that hope I throw mine eyes to heaven, Scorning this base earth and all its toys."
      ],
    "down":
     ["\nThe lowest trees have tops, the ant her gall, The fly her spleen, the little spark his heat.",
      "\nI am not yet so low but that my nails can reach unto thine eyes.",
      "\nMy words fly up, my thoughts remain below:\nWords without thoughts never to heaven go."
      ]
}

# Ask the user to guees a value
guess = int(input("\nGood sir, I humbly requesteth thy conjecture, betwixt zero and fifty, of a chosen numeral."))

# Check the hint, provide a clue or congratulate
while actual_number != guess:
  
  if actual_number > guess:
    print(random.choice(quotes["up"]))
  
  else:
    print(random.choice(quotes["down"]))

  guess = int(input("\nGood sir, I humbly requesteth thy conjecture, betwixt zero and fifty, of a chosen numeral."))
  
# Congratulate the user
print("\nO wonderful, wonderful, and most wonderful wonderful! And yet again wonderful, and after that out of all whooping!")
```
